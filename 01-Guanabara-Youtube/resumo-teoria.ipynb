## Conteúdo Estudado

### 1. Variáveis, Tipos de Dados e Operações Básicas
- Tipos de dados: `int`, `float`, `str`, `bool`
- Entrada de dados: `input()`
- Conversão de tipos: `int()`, `float()`, `str()`
- Operadores aritméticos: `+`, `-`, `*`, `/`, `//`, `%`, `**`
- Operadores de comparação: `==`, `!=`, `>`, `<`, `>=`, `<=`
- Operadores lógicos: `and`, `or`, `not`

---

### 2. Módulos e Bibliotecas
- Importação de módulos: `import modulo` ou `from modulo import funcao`
- Bibliotecas estudadas:
  - `math`: funções matemáticas como `sqrt`, `ceil`, `floor`, `trunc`, `pow`, `factorial`, `hypot`
  - `random`: geração de números aleatórios (`randint`, `random`, `choice`, `shuffle`)
  - `emoji`: inserção de emojis no terminal
  - `pygame`: reprodução de arquivos de áudio MP3

**Exemplo:**

````python
from math import sqrt, floor
num = 16
print(f"Raiz quadrada de {num}: {floor(sqrt(num))}")
````

### 3. Manipulação de Strings

- Funções úteis:
    - `len():` tamanho da string
    - `strip():` remove espaços extras
    - `split():` divide a string em lista de palavras
    - `count():` conta ocorrências de caracteres
    - `find():` retorna índice da primeira ocorrência
    - `replace():` substitui partes da string
    - `upper()`, `lower()`, `capitalize()`, `title()`: formatações de texto
- Fatiamento: `frase[inicio:fim:passo]`

````python
frase = "Curso em Vídeo Python"
print(frase[9:14])   # Saída: Vídeo
print(frase[:5])     # Saída: Curso
print(frase[9::3])   # Saída: VdoPh
````

### 4. Condições (`If` / `Elif` / `Else`)

Estruturas de decisão para desvio de fluxo:

```python
if condicao_1:
    # Código se a condição 1 for True
elif condicao_2:
    # Código se a condição 1 for False e a 2 for True
else:
    # Código se todas as condições anteriores forem False
```

#### Exemplos Práticos Abordados:

* Verificação de notas, aprovação ou reprovação
* Número par ou ímpar
* Velocidade e multa de trânsito
* Verificação de Triângulo válido
* Cálculo de Aumento salarial
* Ano bissexto

#### Exemplo de Condição:

````python
nota = 7.5
if nota >= 6:
    print("Aprovado")
else:
    print("Reprovado")
````
