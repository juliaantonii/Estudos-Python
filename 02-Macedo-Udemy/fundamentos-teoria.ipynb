## 2. Variáveis e Tipagem de Dados
Python possui **tipagem dinâmica**, o que significa que você não precisa declarar o tipo da variável explicitamente; o interpretador identifica o tipo pelo valor atribuído.

### Tipos Primitivos:
* `str`: Cadeias de caracteres (textos).
* `int`: Números inteiros.
* `float`: Números decimais (ponto flutuante).
* `bool`: Valores lógicos (True/False).

**Regras de Nomeclatura:** Variáveis devem ser claras, não podem começar com números e seguem o padrão `snake_case` (ex: `ano_lancamento`).

```python
# Pythonflix - streaming de filmes em python
# Vale resaltar que, pelas regras de sintaxe, as variáveis python não podem::
#   - Iniciar com número
#   - Conter espaços
#   - Conter caracteres especiais
#   - São case-sensitive

name = "Interestellar"
year_launch = 2014
note_movie = 10.0
plan_included = True

# Utilizamos a função type() para descobrirmos o tipo de dado
print(type(name))
print(type(year_launch))
print(type(note_movie))
print(type(plan_included))

# Para visualizarmos:
print(f"\nA informações do filme {name}:")
print(name)
print(year_launch)
print(note_movie)
print(plan_included)
```

## 4. Concatenação e Formatação
Existem várias formas de unir textos e variáveis em Python. A forma mais eficiente e legível é o **F-String** (Formatted Strings), onde inserimos variáveis diretamente dentro do texto usando chaves `{}`.

```python
# Dica: para comentar um trecho de código completo, use ctrl + ;

name = "name"
year_launch = 0
note_movie = 0.0

# name = input("Digite o nome do filme:\n")
# yearLaunch = int(input("Digite o ano de lançamento do filme:\n"))
# noteMovie = float(input("Digite a nota do filme:\n"))

# Alternatica n.01
print("Dados do Filme:")
print("-=-=-=-=-=-=-=-=-=-")
print("Nome do Filme:" ,name)
print("Ano do Filme:" ,year_launch)
print("Nota do Filme:" ,note_movie)
print("\n")

# Alternativa n.02
print("Dados do Filme:", "\n-=-=-=-=-=-=-=-=-=-")
print("Nome do Filme:", name, "\nAno de Lançamento:", year_launch, "\nNota do Filme:", note_movie, "\n")

# Alternativa n.03
print("Dados do Filme:\n"
      "-=-=-=-=-=-=-=-=-=-\n"
      f"Nome do Filme: {name}\n"
      f"Ano do Filme: {year_launch}\n"
      f"Nota do Filme: {note_movie}\n"
      )
```

## 5. Operadores do Sistema

### 5.1 Operadores Aritméticos
Utilizados para cálculos matemáticos. Além dos básicos, lembre-se:
* `%`: Resto da divisão (módulo).
* `**`: Exponenciação (potência).

### 5.2 Operadores de Comparação (Relacionais)
Utilizados para testar condições, resultando sempre em um valor Booleano (`True` ou `False`).

```python
# num1 = int(input("Digite o primeiro numero: \n"))
# num2 = int(input("Digite o segundo numero: \n"))

num1 = 2
num2 = 3

# Operadores Aritméticos
sum = num1 + num2
sub = num1 - num2
div = num1 / num2
mult = num1 * num2
mod = num1 % num2 # resto da divisão
exp = num1 ** num2 # exponenciação

print("Resultados das Operações Aritméticas: \n")
print(f"A soma de {num1} e {num2} é igual a {sum}")
print(f"A subtração de {num1} e {num2} é igual a {sub}")
print(f"A divisão de {num1} e {num2} é igual a {div}")
print(f"A multiplicação de {num1} e {num2} é igual a {mult}")
print(f"O resto da divisão de {num1} e {num2} é igual a {mod}")
print(f"A exponenciação de {num1} e {num2} é igual a {exp}")

print("\n-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-\n")

# Operadores de Comparação
bigger = num1 > num2
smaller = num1 < num2
equal = num1 == num2
different = num1 != num2
bigger_equal = num1 >= num2
smaller_equal = num1 <= num2

print("Resultados das Operações de Comparação: \n")
print(f"O número {num1} é maior que {num2}? {bigger}")
print(f"O número {num1} é menor que {num2}? {smaller}")
print(f"Os números {num1} e {num2} são iguais? {equal}")
print(f"Os números {num1} e {num2} são diferentes? {different}")
print(f"O número {num1} é maior ou igual a {num2}? {bigger_equal}")
print(f"O número {num1} é menor ou igual a {num2}? {smaller_equal}")

# Operadores de Atribuição

num1 += 1 # num1 = num1 + 1
num1 -= 1 # num1 = num1 - 1
num1 *= 1 # num1 = num1 * 1
num1 /= 1 # num1 = num1 / 1
```
