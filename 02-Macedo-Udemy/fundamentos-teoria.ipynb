### 2. Variáveis e Tipagem de Dados .ᐟ.ᐟ
Python possui **tipagem dinâmica**, o que significa que você não precisa declarar o tipo da variável explicitamente; o interpretador identifica o tipo pelo valor atribuído.

#### Tipos Primitivos:
* `str`: Cadeias de caracteres (textos).
* `int`: Números inteiros.
* `float`: Números decimais (ponto flutuante).
* `bool`: Valores lógicos (True/False).

**Regras de Nomeclatura:** Variáveis devem ser claras, não podem começar com números e seguem o padrão `snake_case` (ex: `ano_lancamento`).

```python
# Pythonflix - streaming de filmes em python
# Vale resaltar que, pelas regras de sintaxe, as variáveis python não podem::
#   - Iniciar com número
#   - Conter espaços
#   - Conter caracteres especiais
#   - São case-sensitive

name = "Interestellar"
year_launch = 2014
note_movie = 10.0
plan_included = True

# Utilizamos a função type() para descobrirmos o tipo de dado
print(type(name))
print(type(year_launch))
print(type(note_movie))
print(type(plan_included))

# Para visualizarmos:
print(f"\nA informações do filme {name}:")
print(name)
print(year_launch)
print(note_movie)
print(plan_included)
```

### 5. Operadores do Sistema .ᐟ.ᐟ

#### 5.1 Operadores Aritméticos
Utilizados para cálculos matemáticos. Além dos básicos, lembre-se:
* `%`: Resto da divisão (módulo).
* `**`: Exponenciação (potência).

```python
# num1 = int(input("Digite o primeiro numero: \n"))
# num2 = int(input("Digite o segundo numero: \n"))

num1 = 2
num2 = 3

# Operadores Aritméticos
sum = num1 + num2
sub = num1 - num2
div = num1 / num2
mult = num1 * num2
mod = num1 % num2 # resto da divisão
exp = num1 ** num2 # exponenciação

print("Resultados das Operações Aritméticas: \n")
print(f"A soma de {num1} e {num2} é igual a {sum}")
print(f"A subtração de {num1} e {num2} é igual a {sub}")
print(f"A divisão de {num1} e {num2} é igual a {div}")
print(f"A multiplicação de {num1} e {num2} é igual a {mult}")
print(f"O resto da divisão de {num1} e {num2} é igual a {mod}")
print(f"A exponenciação de {num1} e {num2} é igual a {exp}")
```


#### 5.2 Operadores de Comparação (Relacionais)
Utilizados para testar condições, resultando sempre em um valor Booleano (`True` ou `False`).

```python
# Operadores de Comparação
bigger = num1 > num2
smaller = num1 < num2
equal = num1 == num2
different = num1 != num2
bigger_equal = num1 >= num2
smaller_equal = num1 <= num2

print("Resultados das Operações de Comparação: \n")
print(f"O número {num1} é maior que {num2}? {bigger}")
print(f"O número {num1} é menor que {num2}? {smaller}")
print(f"Os números {num1} e {num2} são iguais? {equal}")
print(f"Os números {num1} e {num2} são diferentes? {different}")
print(f"O número {num1} é maior ou igual a {num2}? {bigger_equal}")
print(f"O número {num1} é menor ou igual a {num2}? {smaller_equal}")
```

#### 5.3 Operadores de Atribuição
Utilizados para definir ou atualizar o valor de uma variável, armazenando o valor do operando à direita no operando à esquerda.

```python
# Operadores de Atribuição

num1 += 1 # num1 = num1 + 1
num1 -= 1 # num1 = num1 - 1
num1 *= 1 # num1 = num1 * 1
num1 /= 1 # num1 = num1 / 1
```

### 6. Utilização de Strings .ᐟ.ᐟ
Strings em Python são **sequências de caracteres** imutáveis. 

* **Case-sensitivity:** Python diferencia maiúsculas de minúsculas (`"Julia" != "julia"`).
* **Multi-line Strings:** Úteis para textos longos ou blocos de documentação, definidas por três aspas (`"""`).
* **Operador `in`:** Permite verificar a existência de um sub-texto dentro de uma string (retorna um Booleano).

```python
movie_name = "Interestelar"
movie_name2 = "interestelar"

print(movie_name == movie_name2)

movie_description = """
    "Interestelar" (2014), dirigido por Christopher Nolan, é uma aclamada
    ficção científica que narra a jornada do ex-piloto Cooper (Matthew McConaughey)
    em busca de um novo lar para a humanidade em uma Terra devastada.
"""

print(movie_name)
print("-"*90)
print(movie_description)

print("Interestelar" in movie_name)
print("interestelar" in movie_name)
print("piloto" in movie_description)
```

A sintaxe básica é `[início : fim : passo]`. 

| Parâmetro | Descrição |
| :--- | :--- |
| **Início** | Índice onde a fatia começa (incluso). |
| **Fim** | Índice onde a fatia termina (**excluso**). |
| **Passo** | Intervalo entre os caracteres (padrão é 1). |

> **Dica:** Use `[::-1]` para inverter uma string de forma rápida e "pythônica"

```python
movie_name = "Gravity Falls"

# string[inicio:fim] - indice começa na posição 0 | indice final - 1

# 1. Buscar toda a string a partir da primeira posição
print(movie_name)
print(movie_name[0:])
print(movie_name[:13])
print("-"*20)

# 2. Buscar toda a string da terceira até a última posição
print(movie_name[2:])
print("-"*20)

# string[inicio:fim:passo] - indice começa na posição 0 | indice final - 1
# O passo é responsável por determinar o incremento. Por padrão esse número é o 1. Por padrão ele faz de 1 em 1.

# 3. Buscar toda a string de 2 em 2 caracteres
print(movie_name[::2])
print("-"*20)

# 4. Buscar toda a string nos índices impares
print(movie_name[1::2]) # o índice 0 é par
print("-"*20)

# 5. Inverter uma string de trás para frente
print(movie_name[::-1])
print("-"*20)
```

### 7. Métodos de String .ᐟ.ᐟ
Os métodos permitem transformar ou analisar strings sem alterar a variável original.

* **Formatação:** `.upper()`, `.lower()`, `.capitalize()`, `.title()`.
* **Busca:** `.find()` (retorna o índice) e `.replace()` (substitui trechos).
* **Divisão:** `.split()` (transforma a string em uma lista baseada em um separador).
* **Alinhamento:** `.center()`, `.ljust()`.

```python
movie_name = "Steve Universe"
movie_description = """
    Steven Universe é uma aclamada série de animação do Cartoon Network, que segue a vida de um menino meio humano e meio Gem, Steven, que vive na cidade fictícia de Beach City com as Crystal Gems.
"""

print(movie_name.upper())
print(movie_name.lower())
print(movie_name.capitalize())
print(movie_name.title())
print(movie_name.center(10, '-'))
print(movie_name.find("i"))
print(movie_name.count("e")) # conta caracter
print(movie_name.replace("Steve", "Julia"))
print(movie_description.split(','))
```
