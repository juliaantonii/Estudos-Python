# ✎ Lista 1: Fundamentos de Variáveis e Tipagem - Parte 1
### 1. Declaração e Verificação
Crie uma variável chamada `idade_base` com o valor 25 e exiba o seu tipo primitivo no terminal.
### 2. Ponto Flutuante
Declare uma variável `preco_unitario` com o valor 19.90 e exiba a classe desse objeto.
### 3. Texto e Aspas
Crie uma variável `identificador` que armazene o texto "ID-99" e exiba o conteúdo em letras maiúsculas.
### 4. Lógica Booleana
Atribua o valor verdadeiro a uma variável `status_ativo` e exiba o resultado da negação (`not`) dessa variável.
### 5. Conversão Estrita
Converta o valor flutuante 15.75 para o tipo inteiro e exiba o resultado final.
### 6. String Numérica
Receba o valor "500" como string, converta para inteiro e realize uma soma de 100 sobre esse valor.
### 7. Saída Formatada
Exiba o conteúdo de três variáveis diferentes (nome, versao, ano) em uma única linha de comando.
### 8. Atribuição Múltipla
Crie três variáveis (a, b, c) com os valores 1, 2 e 3 em uma única linha de código.
### 9. Constantes de Texto
Declare uma variável com uma frase longa e exiba apenas a quantidade total de caracteres presentes nela.
### 10. Limpeza de Memória
Crie uma variável, exiba seu valor e em seguida utilize o comando `del` para removê-la do escopo.
