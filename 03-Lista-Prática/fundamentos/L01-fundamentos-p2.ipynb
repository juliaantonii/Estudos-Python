
# ✎ Lista 2: Fundamentos de Variáveis e Tipagem - Parte 2
### 1. Cálculo de Média
Calcule a média aritmética de quatro valores numéricos inteiros e exiba o resultado com duas casas decimais.
### 2. Conversão de Medidas
Crie um programa que receba um valor em metros e calcule a conversão para milímetros.
### 3. Geometria Plana
Calcule a área de um círculo dado um raio de 5 metros (considere pi = 3.14) e exiba o valor final.
### 4. Manipulação de Strings
Crie uma variável com o nome de um componente de software e exiba as três primeiras letras desse nome.
### 5. Interpolação de Dados
Exiba uma mensagem utilizando f-strings que combine um nome de produto e seu respectivo preço formatado.
### 6. Potenciação
Calcule o valor de 2 elevado à décima potência e exiba o resultado no console.
### 7. Aritmética de Resto
Exiba o resto da divisão de 15 por 4 utilizando o operador de módulo (`%`).
### 8. Entrada de Dados
Solicite que o usuário digite um número e exiba o dobro desse valor, garantindo a conversão correta do tipo.
### 9. Localização de Caracteres
Exiba o índice da primeira ocorrência da letra "a" em uma string pré-definida de sua escolha.
### 10. Divisão Inteira
Realize a divisão de 10 por 3 e exiba apenas a parte inteira do quociente, descartando as casas decimais.
