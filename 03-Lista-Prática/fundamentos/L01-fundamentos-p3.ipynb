# ✎ Lista 3: Fundamentos de Variáveis e Tipagem - Parte 3

### 1. Verificação de Tipo com `isinstance`
Crie uma variável numérica e use a função `isinstance()` para verificar se ela pertence ao tipo `int` ou `float`, exibindo um booleano como resposta.

### 2. Formatação Científica e de Milhar
Declare uma variável com o valor `1500000` (um milhão e meio) e exiba-a formatada com vírgulas separadoras de milhar e também em notação científica.

### 3. Manipulação de Strings (Sanitização)
Dada a string `"  Python 3.12  "`, utilize métodos para remover os espaços em branco nas extremidades e verifique se o resultado final termina com o caractere "2".

### 4. Substituição de Padrões
Crie uma variável com a frase "O custo é R$ 10,00". Utilize o método `.replace()` para trocar a vírgula por ponto e o símbolo "R$" por "USD".

### 5. F-Strings Avançadas (Alinhamento)
Exiba o valor de uma variável `item = "Teclado"` dentro de uma f-string com 20 caracteres de preenchimento, centralizando o texto entre símbolos de hífen (`---Teclado---`).

### 6. Desempacotamento de Sequências
Dada a tupla `dados = (2024, "Produção", "v1.0")`, realize o desempacotamento para três variáveis distintas em uma única linha e exiba apenas a versão.

### 7. Verificação de Conteúdo Numérico
Receba uma entrada do usuário e utilize o método `.isdigit()` para verificar se o que foi digitado pode ser convertido para inteiro antes de realizar qualquer cálculo.

### 8. Atribuição Aritmética Compacta
Declare `x = 10`. Utilize operadores de atribuição composta (`+=`, `*=`, etc.) para somar 5, multiplicar por 2 e exibir o valor final sem reatribuir manualmente.

### 9. Representação de Objeto (`repr` vs `str`)
Crie uma variável de texto e exiba a diferença entre as funções `print(str(variavel))` e `print(repr(variavel))`, observando como o Python enxerga as aspas internas.

### 10. Tipagem Explícita (Type Hints)
Escreva uma linha de código que declare uma variável `limite: int = 100`. Embora o Python não obrigue o tipo, explique em um comentário por que essa prática é útil em projetos profissionais.
