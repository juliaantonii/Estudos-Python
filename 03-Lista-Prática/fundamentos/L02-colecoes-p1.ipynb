# ✎ Lista 4: Coleções (Listas e Tuplas) - Parte 1
### 1. Inicialização de Lista
Crie uma lista com 5 nomes de sistemas operacionais e exiba a lista completa no terminal.

### 2. Acesso Direto
Exiba o elemento localizado no índice 2 de uma lista contendo 6 valores numéricos.

### 3. Indexação Negativa
Acesse e exiba o último elemento de uma lista sem utilizar a função `len()`.

### 4. Atualização de Registro
Altere o valor do terceiro elemento de uma lista de marcas de hardware para "Genérico".

### 5. Expansão de Lista
Adicione o item "Python" ao final de uma lista de tecnologias existente.

### 6. Inserção Específica
Insira o valor 999 na posição de índice 1 de uma lista de IDs numéricos.

### 7. Remoção por Índice
Remova o último elemento de uma lista utilizando o método `.pop()` e exiba o item removido.

### 8. Remoção por Valor
Localize e remova o elemento "Legado" de uma lista de módulos de software.

### 9. Criação de Tupla
Declare uma tupla contendo três siglas de estados e tente exibir apenas a segunda sigla.

### 10. Verificação de Extensão
Exiba a quantidade total de itens presentes em uma lista de bibliotecas de código.
