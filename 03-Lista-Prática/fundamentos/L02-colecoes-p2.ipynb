# ✎ Lista 5: Coleções (Listas e Tuplas) - Parte 2

### 1. Recorte de Coleção
Dada uma lista de 1 a 10, exiba apenas os itens do índice 0 ao índice 4 utilizando slicing.

### 2. Slicing Intervalado
Exiba apenas os elementos de índices ímpares de uma lista contendo 20 números inteiros.

### 3. Inversão de Ordem
Exiba todos os elementos de uma lista em ordem inversa utilizando a técnica de fatiamento.

### 4. Concatenação de Listas
Crie duas listas distintas e gere uma terceira lista que seja a junção das duas primeiras.

### 5. Replicação de Dados
Crie uma lista com o elemento [0] e exiba uma nova lista que repita esse elemento 50 vezes.

### 6. Busca de Posição
Exiba o índice exato onde o valor "Erro" se encontra em uma lista de logs.

### 7. Contagem de Frequência
Conte quantas vezes o número 10 aparece em uma lista de resultados de testes.

### 8. Ordenação Crescente
Ordene uma lista de preços desordenados em ordem crescente e exiba o resultado.

### 9. Matriz Simples
Crie uma lista aninhada que represente uma matriz 2x2 e exiba o elemento da segunda linha e primeira coluna.

### 10. Transformação de Coleção
Converta uma lista de 5 elementos em uma tupla imutável e exiba o novo tipo do objeto.
