# ✎ Lista 6: Coleções (Listas e Tuplas) - Parte 3

### 1. Cópia Rasa vs. Profunda (Shallow vs Deep Copy)
Dada uma lista `original = [1, [2, 3]]`, crie uma cópia que, ao alterar o elemento interno `[2, 3]`, não modifique a lista original. Explique por que o método `.copy()` não seria suficiente neste caso.

### 2. Ordenação Personalizada com `key`
Crie uma lista de tuplas representando produtos: `[("Teclado", 150), ("Mouse", 80), ("Monitor", 900)]`. Ordene a lista pelo preço (segundo elemento) em ordem decrescente.

### 3. Remoção Segura de Itens
Tente remover um item de uma lista que não existe nela (ex: remover "Java" de uma lista de Python). Utilize uma estrutura condicional ou `try/except` para evitar que o programa pare com um erro.

### 4. Compactação de Dados com `zip`
Dada uma lista de `nomes` e uma lista de `precos`, utilize a função `zip()` para criar uma lista de tuplas combinando os dados correspondentes e exiba o resultado.

### 5. Filtragem Funcional
Utilize a função `filter()` ou uma List Comprehension simples para extrair apenas os números maiores que 100 de uma lista numérica pré-definida.

### 6. Desempacotamento Estendido
Dada uma lista com 10 elementos, utilize o operador `*` para capturar o primeiro elemento em uma variável, o último em outra e todos os elementos do meio em uma terceira variável chamada `miolo`.

### 7. Verificação de Identidade em Coleções
Crie duas listas com valores idênticos. Comprove, utilizando o operador `is`, que embora os valores sejam iguais, elas ocupam lugares diferentes na memória.

### 8. Memória e Imutabilidade
Crie uma tupla com um elemento único (singleton). Exiba o tipo do objeto e verifique se ele realmente se comporta como uma tupla (tente adicionar um item e observe o erro).

### 9. Localização com Critério de Erro
Utilize o método `.index()` para buscar um valor. Caso o valor não exista, capture a exceção `ValueError` e exiba uma mensagem personalizada "Item não localizado no inventário".

### 10. Performance: `list` vs `range`
Crie uma lista de 1 milhão de números usando `list(range(1000000))`. Utilize a função `sum()` para somar todos os valores e observe a velocidade de processamento da coleção.
