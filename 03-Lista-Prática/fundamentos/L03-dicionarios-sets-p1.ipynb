# ✎ Lista 7: Dicionários e Sets - Parte 1

### 1. Estrutura de Chave-Valor
Crie um dicionário que armazene as especificações de um servidor: "cpu", "ram" e "storage".

### 2. Consulta de Atributo
Exiba o valor associado à chave "ram" do dicionário criado anteriormente.

### 3. Inserção Dinâmica
Adicione uma nova chave "ip" com um endereço de rede ao seu dicionário de servidor.

### 4. Sobrescrita de Dados
Atualize o valor da chave "storage" para "1TB" e exiba o dicionário atualizado.

### 5. Exclusão com Retorno
Remova a chave "cpu" utilizando o método `.pop()` e exiba o valor que foi removido.

### 6. Remoção Definitiva
Utilize a palavra-chave `del` para excluir uma entrada específica de um dicionário de configurações.

### 7. Definição de Conjunto
Crie um Set contendo 5 códigos de erro e exiba-o para verificar a ausência de ordem.

### 8. Adição em Set
Inclua um novo código de erro em um Set existente e exiba o resultado.

### 9. Remoção Segura
Remova um elemento de um Set utilizando o método `.discard()` para evitar erros caso o item não exista.

### 10. Esvaziamento de Estrutura
Utilize o método `.clear()` para apagar todos os dados de um dicionário e exiba o objeto vazio.
