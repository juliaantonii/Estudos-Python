

# Lista 6: Dicionários e Sets - Parte 2
### 1. Extração de Chaves
Exiba apenas as chaves presentes em um dicionário de usuários do sistema.
### 2. Extração de Valores
Exiba apenas os valores armazenados em um dicionário de preços de produtos.
### 3. Visualização de Itens
Exiba todos os pares chave-valor de um dicionário no formato de lista de tuplas.
### 4. Teste de Existência
Verifique se a chave "admin" existe em um dicionário de permissões e exiba um booleano.
### 5. Clonagem de Objeto
Crie uma cópia independente de um dicionário utilizando o método `.copy()`.
### 6. União de Conjuntos
Realize a união de dois Sets de IDs de sensores e exiba o resultado único.
### 7. Intersecção de Dados
Exiba apenas os elementos que estão presentes simultaneamente em dois conjuntos diferentes.
### 8. Diferença de Conjuntos
Exiba os elementos que existem no primeiro Set, mas não estão presentes no segundo Set.
### 9. Deduplicação de Lista
Converta uma lista com nomes repetidos em um Set e, em seguida, converta-o de volta para uma lista.
### 10. Mapeamento Direto
Crie um dicionário vinculando uma lista de chaves e uma lista de valores utilizando a função `zip()`.
