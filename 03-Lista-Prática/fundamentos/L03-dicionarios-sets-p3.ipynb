# ✎ Lista 9: Dicionários e Sets - Parte 3

### 1. Busca Segura com `get()`
Dada uma configuração de sistema, utilize o método `.get()` para buscar a chave "debug_mode". Caso ela não exista, o programa deve retornar o valor padrão `False` sem gerar erro.

### 2. Dicionário Aninhado (Estrutura JSON)
Crie um dicionário representando um usuário, contendo chaves como "nome", "email" e uma chave "endereco" que armazena outro dicionário com "cidade" e "estado".

### 3. Atualização em Massa com `update()`
Combine dois dicionários de configurações (um de "padrões" e um de "usuário") em um único objeto utilizando o método `.update()`.

### 4. Remoção do Último Item (`popitem`)
Utilize o método `.popitem()` em um dicionário de logs e explique, em um comentário, qual a regra de remoção (LIFO) aplicada neste método em versões recentes do Python.

### 5. Criação com Valores Padrão (`fromkeys`)
Crie um dicionário de "status_servidores" onde as chaves são uma lista de nomes de máquinas e todos os valores iniciam como "offline" usando o método `fromkeys()`.

### 6. Diferença Simétrica de Sets
Dados dois conjuntos de bibliotecas instaladas em ambientes diferentes, exiba apenas os itens que **não** são comuns a ambos (o que existe em um ou no outro, mas não em ambos).

### 7. Verificação de Subconjunto (`issubset`)
Verifique se um Set de "permissões_atribuídas" é um subconjunto de um Set de "permissões_totais" do sistema.

### 8. Congelamento de Conjunto (`frozenset`)
Crie um `frozenset` com IDs de departamentos. Tente adicionar um novo ID e observe o erro, comentando sobre a imutabilidade dessa estrutura.

### 9. Acesso a Dados Aninhados
Acesse e exiba apenas o nome da rua de um dicionário aninhado (como o criado no exercício 2) utilizando a navegação por múltiplas chaves: `dict['chave1']['chave2']`.

### 10. Performance: Associação em Set vs Lista
Crie uma lista e um set com 1.000.000 de itens. Utilize o operador `in` para buscar um valor em ambos e comente sobre a diferença de velocidade percebida (Complexidade O(1) vs O(n)).
