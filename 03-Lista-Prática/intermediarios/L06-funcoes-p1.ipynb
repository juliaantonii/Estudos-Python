# Lista 11: Funções - Parte 1
### 1. Procedimento de Saudação
Defina uma função que exiba "Iniciando Sistema..." sempre que for chamada.
### 2. Função de Soma
Crie uma função que receba dois parâmetros numéricos e exiba a soma entre eles.
### 3. Retorno de Dobro
Desenvolva uma função que receba um número e retorne (return) o seu valor multiplicado por dois.
### 4. Parâmetro Opcional
Crie uma função de boas-vindas que aceite um nome, mas tenha "Usuário" como valor padrão caso nada seja enviado.
### 5. Verificador de Paridade
Crie uma função que receba um número e retorne `True` se for par e `False` se for ímpar.
### 6. Funções Compostas
Crie uma função que chame outra função internamente para realizar um cálculo de bônus salarial.
### 7. Extração de Lista
Desenvolva uma função que receba uma lista e retorne exclusivamente o primeiro elemento dela.
### 8. Função Vazia
Defina uma função utilizando a palavra-chave `pass` e execute-a sem gerar erros.
### 9. Cálculo de Área
Crie uma função que receba o raio de um círculo e retorne a área calculada (pi * r²).
### 10. Documentação Interna
Crie uma função simples e adicione uma Docstring explicando seu funcionamento; exiba essa explicação usando `help()`.
