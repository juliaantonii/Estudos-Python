

# Lista 12: Funções - Parte 2
### 1. Retorno Múltiplo
Crie uma função que receba dois números e retorne tanto a soma quanto a subtração deles em uma tupla.
### 2. Argumentos Variáveis
Desenvolva uma função que aceite uma quantidade indeterminada de números (`*args`) e exiba a soma de todos.
### 3. Argumentos Nomeados
Crie uma função que receba parâmetros nomeados (`**kwargs`) e exiba cada chave e valor formatados.
### 4. Expressão Lambda
Crie uma função lambda que receba um número e retorne o seu triplo.
### 5. Filtro com Lambda
Utilize a função `filter()` junto com uma lambda para extrair apenas números maiores que 10 de uma lista.
### 6. Mapeamento com Lambda
Utilize a função `map()` e uma lambda para converter uma lista de strings para letras maiúsculas.
### 7. Recursão Simples
Crie uma função recursiva que exiba uma contagem regressiva de um número N até 0.
### 8. Teste de Escopo
Crie uma variável global e tente alterá-la dentro de uma função usando a palavra-chave `global`.
### 9. Função de Alta Ordem
Desenvolva uma função que receba outra função como argumento e a execute dentro de seu bloco.
### 10. Gerador de Dados
Crie uma função geradora utilizando `yield` que retorne números de 1 a 3 conforme for solicitada.
