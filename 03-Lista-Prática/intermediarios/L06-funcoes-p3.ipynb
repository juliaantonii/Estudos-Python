# ✎ Lista 18: Funções - Parte 3

### 1. Tipagem de Parâmetros (Type Hints)
Crie uma função que calcule a velocidade média. Utilize Type Hints para indicar que os parâmetros devem ser `float` e que o retorno também será `float`.

### 2. Argumentos Keyword-Only
Defina uma função que receba um `nome` e uma `mensagem`, mas force o uso de argumentos nomeados para a mensagem utilizando o caractere `*` na assinatura da função.

### 3. Closure (Funções Fábrica)
Desenvolva uma função `multiplicador(n)` que retorne uma nova função que multiplica qualquer número por `n`. Teste criando uma função `dobrar` e outra `triplicar`.

### 4. Decorador Simples
Crie um decorador chamado `log_execucao` que exiba "Executando função..." antes de qualquer função que ele envolva ser chamada.

### 5. Memoização Básica
Crie uma função que simule um cálculo pesado e utilize o decorador `@lru_cache` do módulo `functools` para otimizar chamadas repetidas com os mesmos argumentos.

### 6. Desempacotamento de Dicionário em Funções
Crie uma função que aceite três parâmetros (nome, cargo, salario). Defina um dicionário com essas chaves e passe-o para a função utilizando o operador `**`.

### 7. Verificação de Callable
Crie uma lógica que receba um objeto e, utilizando a função `callable()`, verifique se ele pode ser executado como uma função antes de tentar chamá-lo.

### 8. Atributos de Função
Sabia que funções são objetos? Crie uma função e atribua um metadado a ela (ex: `minha_funcao.versao = "1.0"`). Exiba esse atributo no terminal.

### 9. Função de Redução (`reduce`)
Importe `reduce` de `functools` e utilize-a junto com uma lambda para multiplicar todos os elementos de uma lista, obtendo um único valor total.

### 10. Recursão com Proteção
Crie uma função para calcular o Fatorial, mas adicione uma verificação inicial para garantir que o número seja inteiro e positivo, evitando um estouro de pilha (`RecursionError`).
