# ✎ Lista 13: Erros e Módulos - Parte 1
### 1. Tratamento de Divisão
Crie um bloco `try/except` que capture o erro de divisão por zero e exiba uma mensagem de alerta.
### 2. Validação de Tipo
Capture um erro de `ValueError` ao tentar converter uma string não numérica para inteiro.
### 3. Fluxo de Sucesso
Utilize a cláusula `else` em um bloco de tratamento de erros para exibir "Operação realizada com sucesso".
### 4. Finalização de Bloco
Utilize a cláusula `finally` para exibir "Fim do processo" independente de ter ocorrido erro ou não.
### 5. Módulo Matemático
Importe o módulo `math` e exiba a raiz quadrada de 144.
### 6. Manipulação de Data
Importe o módulo `datetime` e exiba a data e hora atual do sistema.
### 7. Geração Aleatória
Utilize o módulo `random` para gerar e exibir um número inteiro aleatório entre 1 e 100.
### 8. Apelido de Módulo
Importe o módulo `os` com o apelido `sistema` e exiba o diretório de trabalho atual.
### 9. Importação Seletiva
Importe apenas a função `choice` do módulo `random` e utilize-a para selecionar um item de uma lista.
### 10. Inspeção de Atributos
Utilize a função `dir()` para listar todos os métodos disponíveis dentro do módulo `math`.
