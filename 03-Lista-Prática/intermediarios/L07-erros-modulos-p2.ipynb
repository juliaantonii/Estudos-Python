
# ✎ Lista 20: Erros e Módulos - Parte 2

### 1. Módulo Externo
Crie um arquivo `.py` com uma função simples, importe-o em outro arquivo e execute a função.

### 2. Captura de IO
Trate o erro de tentativa de abertura de um arquivo que não existe (`FileNotFoundError`).

### 3. Lançamento de Erro
Utilize o comando `raise` para gerar propositalmente um erro caso uma condição de segurança não seja atendida.

### 4. Escrita em Disco
Crie um novo arquivo de texto e escreva três linhas de código Python dentro dele usando o modo 'w'.

### 5. Leitura de Disco
Abra o arquivo de texto criado anteriormente e exiba todo o seu conteúdo no terminal.

### 6. Anexação de Dados
Abra um arquivo existente no modo 'a' (append) e adicione uma nova linha de texto ao final.

### 7. Context Manager
Utilize a instrução `with` para abrir um arquivo e garantir que ele seja fechado automaticamente.

### 8. Leitura Linha a Linha
Percorra um arquivo de texto utilizando um loop e exiba cada linha individualmente.

### 9. Exceção Genérica
Capture qualquer erro inesperado utilizando `Exception as e` e exiba a descrição técnica do erro.

### 10. Verificação de Caminho
Utilize o módulo `os.path` para verificar se um arquivo específico existe no diretório antes de tentar abri-lo.
