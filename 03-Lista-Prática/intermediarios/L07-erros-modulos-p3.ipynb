# ✎ Lista 21: Erros e Módulos - Parte 3

### 1. Criação de Exceção Personalizada
Crie uma classe de erro chamada `SaldoInsuficienteError` que herda de `Exception`. Utilize `raise` para lançar esse erro em uma função de saque bancário simulada.

### 2. Tratamento de Múltiplas Exceções
Escreva um bloco `try` que possa gerar tanto um `IndexError` quanto um `KeyError`. Capture-os de forma distinta, exibindo mensagens específicas para cada um em blocos `except` separados.

### 3. Hierarquia de Exceções
Capture uma exceção e utilize o atributo `__class__.__name__` para exibir o nome exato da classe do erro que foi disparado pelo sistema.

### 4. Criação de Pacotes (`__init__.py`)
Organize dois módulos dentro de uma pasta chamada `utilitarios`. Crie um arquivo `__init__.py` e importe uma função desse pacote em seu script principal para testar a estrutura de pacotes.

### 5. Execução Principal com `__name__ == "__main__"`
Crie um módulo que contenha códigos de teste. Utilize o bloco `if __name__ == "__main__":` para garantir que esses testes só rodem se o arquivo for executado diretamente, e não quando for importado.

### 6. Manipulação de JSON
Importe o módulo `json`. Converta um dicionário Python em uma string JSON e, em seguida, salve essa string em um arquivo com extensão `.json`.

### 7. Temporizador de Execução
Utilize o módulo `time` para medir quantos segundos uma operação de laço (ex: 1 milhão de iterações) leva para ser concluída, exibindo o tempo decorrido.

### 8. Variáveis de Ambiente Simuladas
Utilize o módulo `os` para tentar acessar uma variável de ambiente chamada "USER_AUTH". Caso não exista, capture a falha e atribua um valor padrão de "Convidado".

### 9. Captura de Erros de Importação
Tente importar um módulo que não existe (ex: `import modulo_fantasma`) e capture a exceção `ImportError` para sugerir ao usuário a instalação da dependência necessária.

### 10. Escrita Segura com `try-finally`
Mesmo usando `with`, simule um processo de escrita em arquivo onde, no bloco `finally`, você exibe uma confirmação de que os recursos do sistema foram liberados.
