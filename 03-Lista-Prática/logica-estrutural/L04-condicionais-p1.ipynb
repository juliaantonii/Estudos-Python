# ✎ Lista 10: Condicionais (If/Else) - Parte 1

### 1. Comparação Numérica
Receba um número e exiba se ele é maior que 100 ou se é menor/igual a 100.

### 2. Validação de Sinal
Exiba se um valor numérico inserido é "Positivo", "Negativo" ou "Zero".

### 3. Teste de Paridade
Crie um programa que exiba se um número inteiro é "Par" ou "Ímpar".

### 4. Diferenciação de Valores
Receba dois números e exiba qual deles é o maior ou se ambos são iguais.

### 5. Verificação de Texto
Exiba se uma string inserida pelo usuário está vazia ou se contém caracteres.

### 6. Filtro de Idade
Verifique se uma idade é maior ou igual a 18 e exiba "Acesso Permitido" ou "Bloqueado".

### 7. Teste de Igualdade
Receba duas palavras e exiba se elas são exatamente iguais em termos de caracteres.

### 8. Operador de Diferença
Verifique se o valor de uma variável é diferente de zero e exiba o resultado lógico.

### 9. Intervalo Fechado
Exiba se um número está dentro do intervalo entre 10 e 20 (inclusive).

### 10. Estado Lógico
Exiba "Ativo" se uma variável booleana for verdadeira, caso contrário, exiba "Inativo".
