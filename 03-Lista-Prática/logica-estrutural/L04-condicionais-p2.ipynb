# ✎ Lista 11: Condicionais (If/Else) - Parte 2

### 1. Escala de Notas
Classifique uma nota de 0 a 10 em: "Excelente" (>=9), "Bom" (>=7), "Regular" (>=5) ou "Insuficiente" (<5).

### 2. Menu de Operações
Crie uma calculadora que receba dois números e uma opção (1-Soma, 2-Subtração, 3-Multiplicação) e exiba o cálculo.

### 3. Regra de Ano Bissexto
Determine se um ano inserido é bissexto seguindo as regras do calendário gregoriano.

### 4. Máximo de Três
Receba três valores numéricos e exiba exclusivamente o maior entre eles.

### 5. Autenticação Simples
Simule um sistema onde o acesso só é concedido se o usuário for "admin" e a senha for "1234".

### 6. Tipo de Caractere
Receba uma letra e exiba se ela é uma "Vogal" ou uma "Consoante".

### 7. Classificação de Triângulos
Receba três lados e exiba se o triângulo é "Equilátero", "Isósceles" ou "Escaleno".

### 8. Faixa Etária Técnica
Classifique um usuário como "Menor", "Adulto" (18-64) ou "Sênior" (>=65).

### 9. Cálculo de IMC
Calcule o IMC e exiba a classificação: "Baixo peso", "Peso normal", "Sobrepeso" ou "Obesidade".

### 10. Ternário de Status
Utilize uma única linha de código para atribuir "Aprovado" se a média for >= 7, senão "Reprovado", e exiba o resultado.
