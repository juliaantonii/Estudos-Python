# ✎ Lista 12: Condicionais (If/Else) - Parte 3

### 1. Estrutura Match Case (Python 3.10+)
Crie um sistema de status de pedido que receba um código (ex: "PEN", "CAN", "ENT") e utilize o `match case` para exibir mensagens para "Pendente", "Cancelado" ou "Entregue".

### 2. Validação de Curto-Circuito
Crie uma condição que verifique se uma variável `divisor` é diferente de zero antes de realizar uma divisão, utilizando a lógica de curto-circuito em uma única linha de `if`.

### 3. Operador de Identidade vs. Igualdade
Dada uma variável `x = None`, faça uma verificação profissional utilizando `if x is None:` e explique em um comentário por que usamos `is` em vez de `==` para valores nulos.

### 4. Condicionais com Operador `in` em Dicionários
Verifique se uma chave existe em um dicionário de configurações. Se existir, verifique se o valor é verdadeiro. Exiba "Habilitado" apenas se ambas as condições forem atendidas.

### 5. Truthy e Falsy
Escreva um programa que receba uma lista. Se a lista contiver itens (for "True"), exiba a quantidade; se estiver vazia (for "False"), exiba "Lista Vazia", sem usar explicitamente `len(lista) > 0`.

### 6. Tratamento de Exceções Lógicas
Ao receber dois números para divisão, utilize um bloco `if/else` para evitar o erro `ZeroDivisionError`, exibindo uma mensagem personalizada de aviso ao usuário.

### 7. Verificação de Intervalo com Operadores Encadeados
Em Python, podemos escrever `10 <= x <= 20`. Crie um código que utilize essa sintaxe para validar se uma temperatura está em um intervalo seguro.

### 8. Lógica Booleana Inversa com `not`
Crie um sistema de permissão onde o acesso é negado apenas se o usuário **não** estiver na lista de "banidos" e **não** for menor de idade.

### 9. Avaliação de Múltiplas Flags
Dadas três variáveis booleanas (`tem_estoque`, `cartao_validado`, `cupom_ativo`), crie uma lógica que permita a compra apenas se as duas primeiras forem verdadeiras, independentemente do cupom.

### 10. Ternários Aninhados
Embora devam ser usados com cautela, crie um operador ternário que retorne "Positivo", "Negativo" ou "Zero" em uma única linha de atribuição para uma variável `resultado`.
