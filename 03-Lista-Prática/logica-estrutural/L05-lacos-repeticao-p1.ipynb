# Lista 9: Laços de Repetição (For/While) - Parte 1
### 1. Contador Crescente
Utilize um loop `for` para exibir todos os números de 0 a 10.
### 2. Contador Regressivo
Utilize um loop `for` e a função `range()` para exibir uma contagem regressiva de 10 até 0.
### 3. Iteração de Lista
Percorra uma lista de 5 nomes de ferramentas e exiba cada nome individualmente.
### 4. Decomposição de Texto
Percorra uma string e exiba cada caractere em uma linha diferente.
### 5. Acumulador Numérico
Calcule a soma de todos os números de 1 a 50 utilizando um loop e exiba o total final.
### 6. Loop de Condição
Utilize um `while` para exibir os números de 1 a 5, incrementando a variável de controle manualmente.
### 7. Interrupção Forçada
Crie um loop que percorra de 1 a 10, mas utilize o `break` para parar no número 6.
### 8. Salto de Instrução
Crie um loop de 1 a 10 que utilize o `continue` para não exibir o número 5.
### 9. Tabuada Simples
Receba um número do usuário e exiba a tabuada de 1 a 10 para esse valor.
### 10. Indexação em Loop
Utilize a função `enumerate()` para exibir o índice e o valor de cada item em uma lista de linguagens.
