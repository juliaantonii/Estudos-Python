# ✎ Lista 10: Laços de Repetição (For/While) - Parte 2
### 1. Compreensão de Quadrados
Crie uma lista contendo o quadrado dos números de 1 a 10 utilizando List Comprehension.
### 2. Filtro em Compreensão
Gere uma lista apenas com os números pares entre 0 e 20 utilizando uma única linha de código.
### 3. Estrutura Aninhada
Utilize dois loops `for` (um dentro do outro) para exibir uma grade de coordenadas (x, y) de 0 a 2.
### 4. Busca de Máximo Manual
Encontre e exiba o maior valor de uma lista numérica sem utilizar a função nativa `max()`.
### 5. Reversão de Texto
Receba uma palavra e exiba-a de trás para frente utilizando um laço de repetição.
### 6. Analisador de Vogais
Conte e exiba quantas vogais existem em uma frase fornecida pelo usuário.
### 7. Sequência de Fibonacci
Exiba os 10 primeiros números da sequência de Fibonacci utilizando um laço `while`.
### 8. Fatorial Numérico
Calcule o fatorial de um número inteiro fornecido pelo usuário utilizando um loop.
### 9. Validação de Entrada
Solicite uma senha ao usuário e repita a solicitação enquanto a entrada for diferente de "confirmado".
### 10. Mapeamento por Compreensão
Crie um dicionário onde as chaves são números de 1 a 5 e os valores são seus cubos, usando Dictionary Comprehension.
