# ✎ Lista 15: Laços de Repetição (For/While) - Parte 3

### 1. Cláusula `else` em Loops
Crie um loop `for` que procure por um número negativo em uma lista. Se encontrar, exiba-o e pare com `break`. Se terminar a lista sem encontrar, utilize o `else` do loop para exibir "Nenhum número negativo localizado".

### 2. Simulação de Menu Infinito
Crie um loop `while True` que exiba um menu de opções (1. Iniciar, 2. Ajuda, 3. Sair). O loop só deve ser interrompido se o usuário digitar "3", tratando qualquer outra entrada inválida com uma mensagem de erro.

### 3. Iteração em Dicionários Complexos
Dada uma lista de dicionários (ex: uma lista de produtos com 'nome' e 'preco'), utilize um laço para exibir apenas os nomes dos produtos que custam mais de R$ 100,00.

### 4. Zip e Iteração Paralela
Utilize a função `zip()` dentro de um loop `for` para iterar simultaneamente sobre uma lista de `alunos` e uma lista de `notas`, exibindo frases como: "O aluno [nome] tirou nota [valor]".

### 5. Nested List Comprehension
Dada uma matriz (lista de listas) `[[1, 2], [3, 4], [5, 6]]`, utilize uma única linha de List Comprehension para "achatar" a estrutura em uma lista simples: `[1, 2, 3, 4, 5, 6]`.

### 6. Controle de Tentativas com Limitador
Crie um laço `while` que permita ao usuário tentar adivinhar um número secreto apenas 3 vezes. Ao final das tentativas, informe se ele venceu ou se as chances acabaram.

### 7. Uso do Módulo `itertools` (Básico)
Importe `cycle` do módulo `itertools` para criar um loop que alterne infinitamente entre as cores "Verde", "Amarelo" e "Vermelho", mas utilize um contador interno para interromper o processo após 6 trocas.

### 8. Filtro Condicional com `any()`
Utilize a função `any()` combinada com uma expressão geradora dentro de um `if` para verificar se existe algum número ímpar em uma lista, exibindo uma mensagem de confirmação.

### 9. Manipulação de Iteradores com `next()`
Transforme uma lista de nomes em um iterador usando `iter()` e utilize a função `next()` dentro de um bloco `try/except` para exibir os itens um a um até que a exceção `StopIteration` seja capturada.

### 10. Performance: For Loop vs Map
Crie uma lista de números e utilize a função `map()` com uma função `lambda` para converter todos os números em strings. Compare visualmente a simplicidade deste método com um laço `for` tradicional.
