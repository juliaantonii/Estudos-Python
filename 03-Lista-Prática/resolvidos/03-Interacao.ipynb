# ✎ᝰ Lista 3: Interação (Comandos de Entrada)

### 1. Leia um nome com input().


```python
nome = input("Digite um nome: ")
print(f"Você possui nome: {nome}")
```

### 2. Leia uma idade e converta para int().


```python
idade = int(input("Digite uma idade: "))
print(f"Você possui idade: {idade}")
```

### 3. Leia uma altura e converta para float().


```python
altura = float(input("Digite uma altura: "))
print(f"Você possui altura: {altura}")
```

### 4. Peça ao usuário um número e exiba o dobro dele.


```python
num = int(input("Digite um valor: "))
print(f"O dobro desse valor é: {num*2}")
```

### 5. Peça o nome de um país e exiba "Você quer visitar [país]".


```python
pais = input("Digite o nome de um pais: ")
print(f"Você quer visitar {pais}")
```

### 6. Leia dois números inteiros e exiba a soma.


```python
num1 = int(input("Numero 1: "))
num2 = int(input("Numero 2: "))
print(f"A soma destes é: {num1+num2}")
```

### 7. Leia um valor em Reais e converta para float.


```python
Reais = float(input("Digite um valor em R$: "))
print(f"Você possui: R${Reais}")
```

### 8. Peça uma cor e armazene na variável cor_favorita.


```python
cor = input("Digite sua cor favorita: ")
print(f"Sua cor favorita é: {cor}")
```

### 9. Peça um animal e exiba "O [animal] faz barulho".


```python
animal = input("Digite o nome de um animal: ")
print(f"O {animal} faz barulho")
```

### 10. Peça o ano de nascimento e calcule a idade atual (considere 2026).

```python
ano = int(input("Digite o ano de nascimento: "))
print(f"Você possui {2026-ano} anos de idade")
```
